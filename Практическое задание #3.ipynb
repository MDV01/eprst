```python
import pandas as pd
import re

#Считайте датасет из файла train.csv (это данные о выживаемости на Титанике) - 1 балл
df = pd.read_csv('train.csv')
df
```

```python
#Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д. - 1 балл
df.info()
df.describe()
```

```python
#Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) - 1 балл
df.groupby('Pclass')['Survived'].agg([
    ('total_passengers', 'count'),
    ('survived_count', 'sum'),
    ('survival_rate', lambda x: (x.sum() / x.count() * 100).round(2))
])
```

```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле - 1 балл
def extract_first_name(full_name):
    # всё до запятой
    name_part = full_name.split(',')[1] if ',' in full_name else full_name
    # для замужних женщин
    bracket = re.search(r'\(([^)]+)\)', name_part)
    if bracket:
        name_in_brackets = bracket.group(1)
        # Берем первое слово из скбк
        return name_in_brackets.split()[0]
    # по регулярке Mr., Mrs., Miss, Ms., Master, Dr. и т.д.
    name_match = re.search(r'(?:Mr|Mrs|Miss|Ms|Master|Dr|Col|Major|Capt|Don|Dona|Countess|Mme|Mlle|Lady|Sir)\.?\s+([A-Za-z]+)', name_part)
    if name_match:
        return name_match.group(1)
    words = name_part.split()
    if len(words) >= 2:
        return words[1]
    return full_name

df['FirstName'] = df['Name'].apply(extract_first_name)

# популярные имена по полу
male_names = df[df['Sex'] == 'male']['FirstName'].value_counts()
female_names = df[df['Sex'] == 'female']['FirstName'].value_counts()

print("Самые популярные имена на Титанике:")
print("=" * 50)
print(f"Мужское имя: '{male_names.index[0]}' (встречается {male_names.iloc[0]} раз)")
print(f"Женское имя: '{female_names.index[0]}' (встречается {female_names.iloc[0]} раз)")
```

```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе - 2 балла
popular_names_by_class = df.groupby(['Pclass', 'Sex'])['FirstName'].agg(lambda x: x.value_counts().index[0]).reset_index()
print(popular_names_by_class)
```

```python
#Выведите часть таблицы с пассажирами, возраст которых больше 44 лет - 1 балл
columns_to_show = ['PassengerId', 'Name', 'Age', 'Sex', 'Pclass', 'Survived', 'SibSp', 'Parch', 'Fare']
pass_over_44 = df[df['Age'] > 44][columns_to_show]
print(pass_over_44.to_string(index=False))
```

```python
#Выведите часть таблицы с пассажирами, возраст которых меньше 44 лет и которые мужского пола - 2 балла
male_pass_under_44 = df[(df['Age'] < 44) & (df['Sex'] == 'male')][columns_to_show]
print(male_pass_under_44.to_string(index=False))
```

```python
#Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека) - 1 балл
cabin_occ = df.groupby('Cabin').size().reset_index(name='total_passeng_in_cabin')
cabin_over_2 = cabin_occ[cabin_occ['total_passeng_in_cabin'] > 1]
print(cabin_over_2)
```
