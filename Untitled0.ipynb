```python
class Chocolate:
    def __init__(self, price: float, add: str = None):
        if price <= 0:
            raise ValueError("Цена батончика должна быть больше нуля")


        self.price = price
        self.add = add

    def show_my_chocolate(self):
        if self.add:
            print(f"Шоколадный батончик, добавка: {self.add}")
        else:
            print("Обычный шоколадный батончик")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"

try:
    choco1 = Chocolate(45.50)
    choco1.show_my_chocolate()
    print(choco1)
    print()

    choco2 = Chocolate(65.80, "орехи")
    choco2.show_my_chocolate()
    print(choco2)
    print()

    choco3 = Chocolate(55.30, "изюм")
    choco3.show_my_chocolate()
    print(choco3)
    print()

    print("Попытка создать батончик с отрицательной ценой:")
    choco5 = Chocolate(-10.0, "карамель")

except ValueError as e:
    print(f"Ошибка: {e}")
    print()

try:
        print("Попытка создать батончик с 0:")
        choco6 = Chocolate(0, "карамель")

except ValueError as e:
        print(f"Ошибка: {e}")
        print()
```
