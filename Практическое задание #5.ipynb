```python
#Считайте датасет из файла train.csv (это данные о выживаемости на Титанике) с помощью polars - 1 балл
#
#Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д. - 1 балл
#hint: используйте методы и атрибуты датафрейма: describe(), dtypes, null_count(), mean()...

import polars as pl
df = pl.read_csv("train.csv")
print("Размер:", df.shape)
print("\nТипы данных:")
print(df.dtypes)
print("\nПропуски:")
print(df.null_count())
print("\nОписательные статистики:")
print(df.describe())
print(df.mean())
```

```python
#Посчитайте количество пассажиров каждого класса (Pclass) - 1 балл
#hint: используйте методы: get_column(), value_counts()...
print(df.get_column("Pclass").value_counts().sort("Pclass"))
```

```python
#Выведите количество выживших мужчин и женщин на корабле - 1 балл
#hint: используйте методы: groupby(), agg(), pl.col(), sum()...
survivors_by_sex = df.group_by("Sex").agg(
    pl.col("Survived").sum().alias("survived_count")
).sort("Sex")
print(survivors_by_sex)
```

```python
#Выведите часть таблицы с пассажирами, возраст которых больше 44 лет - 1 балл
#hint: используйте методы: filter(), pl.col()...
older_passengers = df.filter(pl.col("Age") > 44)
print(older_passengers)
```

```python
#Считайте датасет из файла train.csv (это данные о выживаемости на Титанике) с помощью pandas
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
#Посчитайте средний возраст пассажиров и его стандартное отклонение с помощью bottleneck - 1 бал
#hint: используйте методы: bn.nanmean(), bn.nanstd()
import bottleneck as bn
age_array = df['Age'].values
mean_age = bn.nanmean(age_array)
std_age = bn.nanstd(age_array)

print(f"Средний возраст пассажиров: {mean_age:.2f} лет")
print(f"Стандартное отклонение: {std_age:.2f} лет")
```

```python
#Для каждого пассажира умножьте значение столбца Fare на 1.3 и сохраните результаты как новый столбец Fare_new - 1 балл
#hint: используйте методы: .itertuples() или apply()

fare_list = [row.Fare * 1.3 for row in df.itertuples()]
df['Fare_new'] = pd.Series(fare_list).apply(lambda x: round(x, 1))
print(df[['PassengerId', 'Fare', 'Fare_new']].head())
```

```python
#2. Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных - напишите свои выводы в комментариях - 2 балла
for col in df.columns:
    dtype = df[col].dtype
    unique = df[col].nunique() if dtype == 'object' else None
    min_val = df[col].min() if dtype != 'object' else None
    max_val = df[col].max() if dtype != 'object' else None

    print(f"\n{col}:")
    print(f"  Текущий тип: {dtype}")

    if dtype == 'object':
        if unique < len(df) * 0.5:
            print(f"Рекомендация: 'category'")
        else:
            print(f"Оставить 'object'")
    elif 'int' in str(dtype):
        if max_val < 256:
            print(f"Рекомендация: 'int8'")
        elif max_val < 65536:
            print(f"Рекомендация: 'int16'")
    elif 'float' in str(dtype):
        print(f"Для точности оставить float64")

#столбцы с 'yes/no' или ограниченным набором значений могут быть преобразованы в тип 'category':
#mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea - бинарные признаки
#furnishingstatus - 3 категории
#Экономия памяти: 70-90%

#Колонки имеют ограниченный диапазон и могут быть уменьшены:
#bedrooms (2-6) → uint8
#bathrooms (1-4) → uint8
#stories (1-4) → uint8
#parking (0-3) → uint8
#Экономия памяти: 75% (с int64 до uint8)

#Колонки требуют баланса между точностью и памятью: чем выше точность — тем больше памяти потребуется.
#price - лучше оставить float64 для точности
#area - можно использовать float32 или uint16
```

```python
#Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации
df_orig = df.copy()
categorical = ['mainroad', 'guestroom', 'basement', 'hotwaterheating',
               'airconditioning', 'prefarea', 'furnishingstatus']
for col in categorical:
    df[col] = df[col].astype('category')

df['bedrooms'] = df['bedrooms'].astype('uint8')
df['bathrooms'] = df['bathrooms'].astype('uint8')
df['stories'] = df['stories'].astype('uint8')
df['parking'] = df['parking'].astype('uint8')
#df['area'] = df['area'].astype('int16')

before = df_orig.memory_usage(deep=True).sum()
after = df.memory_usage(deep=True).sum()

print(f"До: {before/1024:.1f} KB")
print(f"После: {after/1024:.1f} KB")
print(f"Экономия: {(before-after)/1024:.1f} KB ({(1-after/before)*100:.1f}%)")
```
